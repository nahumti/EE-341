```python
#Import Cell
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage, misc
from skimage import color, transform
```

## Task 1 - Displaying Image

```python
#Task 1
image = plt.imread('DailyShow.jpg')

plt.figure()
plt.imshow(image)
plt.show()

image_gray = color.rgb2gray(image)

plt.figure()
plt.imshow(image_gray, cmap = 'gray')
plt.show()
```

```python
#Task 2
# result = ndimage.filters.convolve(image, kernel)

h1_kernel = [[-1, 0, 1],
             [-2, 0, 2],
             [-1, 0, 1]]

h1_result = ndimage.filters.convolve(image_gray, h1_kernel)
plt.imshow(h1_result, cmap = 'gray')
plt.show()

h2_kernel = [[-1, -2, -1],
             [0, 0, 0],
             [1, 2 , 1]]

h2_result = ndimage.filters.convolve(image_gray, h2_kernel)
plt.imshow(h2_result, cmap = 'gray')
plt.show()

plt.imshow(np.sqrt(h1_result**2 + h2_result**2), cmap = 'gray')
plt.show()

#Optional
plt.imshow(-1*np.sqrt(h1_result**2 + h2_result**2), cmap = 'gray')
plt.show()
```

## Assignment 2 - Edge Detection

```python
#My Own Image
my_image = plt.imread('Sunset.png')
my_gray = color.rgb2gray(my_image) 

h_image = ndimage.filters.convolve(my_gray, h1_kernel)
plt.imshow(h_image, cmap = 'gray')
plt.show()

v_image = ndimage.filters.convolve(my_gray, h2_kernel)
plt.imshow(v_image, cmap = 'gray')
plt.show()

plt.imshow(np.sqrt(h_image**2 + v_image**2), cmap = 'gray')
plt.show()
```

## Assignment 3 - Downscaling

```python
def pick_Scaler(S):
    image = plt.imread('DailyShow.jpg')
    image_gray = color.rgb2gray(image)
    scaled_image = image_gray[S//2::S, S//2::S]
    plt.imshow(scaled_image, cmap = 'gray')
    plt.show()
 

pick_Scaler(2)
pick_Scaler(5)
        
def avg_Scaler(S):
    image = plt.imread('DailyShow.jpg')
    image_gray = color.rgb2gray(image)
    height, width = image_gray.shape
    scaled_height = int(height / S)
    scaled_width = int(width / S)
    scaled_image = np.zeros((scaled_height, scaled_width))
    for i in range(scaled_height):
        for j in range(scaled_width):
            scaled_image[i, j] = image_gray[i*S:(i+1)*S, j*S:(j+1)*S].mean()
    plt.imshow(scaled_image, cmap = 'gray')
    plt.show()        
        
avg_Scaler(2)
avg_Scaler(5)
                
```

## Assignment 4 - Flipping

The first equation X[N-n+1,m] corresponds to the flipping up and down or the numpy.ud function

The second equation X[n, M-m+1] corresponds to the flipping left and right or the numpy.lr function

The third and final which includes both is the combination of numpy.ud and numpy.lr

```python
image = plt.imread('DailyShow.jpg')
flipped_i1 = np.flipud(image)
plt.imshow(flipped_i1)
plt.show()

flipped_i2 = np.fliplr(image)
plt.imshow(flipped_i2)
plt.show()

flipped_i3 = np.fliplr(flipped_i1)
plt.imshow(flipped_i3)
plt.show()
```
